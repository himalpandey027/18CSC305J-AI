```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download necessary NLTK resources
# nltk.download('vader_lexicon')
nltk.download('punkt')

# Initialize the sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Input sentence
sentence = "NLTK is a good platform for building Python programs to work with human language data."

# Tokenize the sentence into words
words = nltk.word_tokenize(sentence)

# Iterate over each word and get sentiment scores
for word in words:
    if word.isalpha():  # Consider only alphabetic words
        scores = sid.polarity_scores(word)
        print(f"Word: {word}")
        print(f"Negative: {scores['neg']}")
        print(f"Neutral: {scores['neu']}")
        print(f"Positive: {scores['pos']}")
        print()
```
